```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('owid-covid-data.csv')
```

```python
df.head()
```

```python
col_del=[]
for i in df.columns:
    if df[i].isna().sum()<len(df)//2:
        col_del.append(i)
```

```python
col_del
```

```python
for i in df.columns:
    if i not in col_del:
        df.drop([i],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop(['new_tests_smoothed','new_tests_smoothed_per_thousand','positive_rate','tests_units','extreme_poverty',
        'new_cases_smoothed','female_smokers','male_smokers'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop(['new_deaths','new_cases_smoothed_per_million','new_deaths_per_million'],axis=1,
       inplace=True)
```

```python
df['continent'].fillna('World',inplace=True)
```

```python
df.drop(['iso_code'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
for i in df.columns:
    if df[i].isna().sum()>0:
        df[i].fillna(df[i].mean(),inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
X=df['new_deaths_smoothed'].values.reshape(-1, 1)
y=df['new_cases'].values.reshape(-1, 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler() 
sc_y = StandardScaler() 
X_std = sc_x.fit_transform(X) 
y_std = sc_y.fit_transform(y) 
lr = LinearRegression() 
lr.fit(X_std, y_std) 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print('mse: %.3f, mae: %.3f' % (mse, mae))
```
